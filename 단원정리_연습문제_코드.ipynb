```python
import pandas as pd
import numpy as np

# 데이터 생성
data = {
    '상품명': ['A', 'B', 'C', 'D', 'E', 'F'],
    '카테고리': ['과일', '과자', '과일', '음료', '과자', '음료'],
    '판매수량': [10, 8, 15, 20, 0, 12],
    '가격': [1500, 2000, np.nan, 1000, 2500, 900]
}

# DataFrame 생성
df = pd.DataFrame(data)

print("원본 데이터:")
print(df)
print("\n" + "="*50 + "\n")

# 1. '판매수량'이 0인 데이터를 결측치(NaN)로 변경
df.loc[df['판매수량'] == 0, '판매수량'] = np.nan

print("1. 판매수량이 0인 데이터를 NaN으로 변경:")
print(df)
print("\n" + "="*50 + "\n")

# 2. '가격' 열의 결측치(NaN)를 '가격' 열의 전체 평균값으로 채우기
price_mean = df['가격'].mean()
df['가격'] = df['가격'].fillna(price_mean)

print(f"2. 가격 열의 평균값: {price_mean:.2f}")
print("   가격 열의 NaN을 평균값으로 채운 결과:")
print(df)
print("\n" + "="*50 + "\n")

# 3. '총판매액' 열을 새로 만들어 계산 (총판매액 = 판매수량 * 가격)
df['총판매액'] = df['판매수량'] * df['가격']

print("3. 총판매액 열 추가 (판매수량 * 가격):")
print(df)
print("\n" + "="*50 + "\n")

# 4. 상품 '카테고리'별 '총판매액'의 평균을 계산하여 출력
category_avg = df.groupby('카테고리')['총판매액'].mean()

print("4. 카테고리별 총판매액 평균:")
print("-" * 30)
for category, avg_sales in category_avg.items():
    if pd.isna(avg_sales):
        print(f"{category}: NaN (판매수량이 0인 상품 포함)")
    else:
        print(f"{category}: {avg_sales:,.2f}원")

print("\n" + "="*50 + "\n")

# 최종 데이터프레임 출력
print("최종 데이터프레임:")
print(df)

# 추가 통계 정보
print("\n" + "="*50 + "\n")
print("추가 통계 정보:")
print("-" * 30)
print(f"전체 판매 상품 수: {len(df)}개")
print(f"결측치가 있는 판매수량: {df['판매수량'].isna().sum()}개")
print(f"전체 총판매액 합계: {df['총판매액'].sum():,.2f}원")
print(f"카테고리별 상품 수:")
print(df.groupby('카테고리').size())
```
